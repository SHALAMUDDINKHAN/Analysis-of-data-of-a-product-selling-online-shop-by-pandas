## Exploratory Data Analysis
### Checking customers who is buying more group by gender Gender


```python
data.columns
```

```python
#count number of customers group by Gender
ax = sns.countplot(x = 'Gender',data= data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#Total amount of selling in Gnders
sales_gen=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=sales_gen)
#for bars in ax.containers:
 #   ax.bar_label(bars)
```

### we can see that most of the buyers are Female from the store

##  Checking customers who is buying more group by age

```python
data.columns
```

```python
#count of  customers group by age
ans=sns.countplot(x='Age Group',hue='Gender',data=data)
for bars in ans.containers:
    ans.bar_label(bars)
```

```python
sales_age=data.groupby(['Age Group'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
#Total amount of  customers group by age
sales_age=data.groupby(['Age Group'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

### the graph is showing that most of the customers are from age 26-35

##  Checking who is buying from respective states

```python
data.columns
```

```python
# total numbers of orders from top 10 state

sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state,x='State', y='Orders')
```

##                                     Total Orders from various states

```python
sales_state
```

```python
# total amount from top 10 state

sales_state=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state,x='State', y='Amount')
```

##                                     Total Amounts from various states

# According to occupation of the customers

```python
sns.set(rc={'figure.figsize':(18,5)})
ax=sns.countplot(data=data,x='Occupation')
for ans in ax.containers:
    ax.bar_label(ans)
```

### the maximum number customers are from IT sectors

```python
sales_Occ=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_Occ,x='Occupation',y='Amount')
```

### the maxcimum amount comes from the  customers workingg in IT sector

# sales according product catagory

```python
sns.set(rc={'figure.figsize':(25,10)})
ax=sns.countplot(data=data,x='Product_Category')
for ans in ax.containers:
    ax.bar_label(ans)
```

## maximum number of poroduct soled from CLOTHING & APPAREL catagory

```python
sales_PC=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_PC,x='Product_Category',y='Amount')
```

## maximum Amount comes  from FOOD catagory

# TOP 10 most sold product  by product ID

```python
sales_PI=data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_PI,x='Product_ID',y='Orders')
```
